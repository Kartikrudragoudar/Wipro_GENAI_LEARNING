```python
import requests
import json
import os
from dotenv import load_dotenv

# ─── Load environment variables from .env file ────────────────────────────────
load_dotenv(override=True)

# ─── Configuration (modify these parameters as needed) ────────────────────────
API_KEY       = os.getenv("ANTHROPIC_API_KEY")
ENDPOINT      = "https://llmgw-wp.tekstac.com/v1/chat/completions"
MODEL         = "global.anthropic.claude-haiku-4-5-20251001-v1:0"
MAX_TOKENS    = 256           # ← modify to change response length
TEMPERATURE   = 0.5           # ← modify between 0.0 (focused) and 1.0 (creative)
SYSTEM_PROMPT = "You are a helpful assistant."  # ← modify to change behavior

# ─── Headers ──────────────────────────────────────────────────────────────────
headers = {
    "x-api-key": API_KEY,
    "anthropic-version": "2023-06-01",
    "content-type": "application/json"
}

# ─── 1. Accept dynamic user input ─────────────────────────────────────────────
user_input = input("You: ").strip()

# ─── Request Body ─────────────────────────────────────────────────────────────
payload = {
    "model": MODEL,
    "max_tokens": MAX_TOKENS,
    "temperature": TEMPERATURE,
    "system": SYSTEM_PROMPT,
    "messages": [
        {
            "role": "user",
            "content": user_input       # ← dynamic input used here
        }
    ]
}

# ─── 2. Send POST Request to LLM endpoint ─────────────────────────────────────
response = requests.post(ENDPOINT, headers=headers, json=payload)

# ─── 3. Extract and print full JSON response ──────────────────────────────────
print(f"\nStatus Code : {response.status_code}")
print(f"\nFull JSON Response:\n{json.dumps(response.json(), indent=2)}")

# ─── Extract and print assistant reply ────────────────────────────────────────
assistant_reply = response.json()["choices"][0]["message"]["content"]
print(f"\nAssistant Reply:\n{assistant_reply}")

# ─── Token usage ──────────────────────────────────────────────────────────────
usage = response.json()["usage"]
print(f"\nTokens Used — Prompt: {usage['prompt_tokens']}, Completion: {usage['completion_tokens']}, Total: {usage['total_tokens']}")
```
