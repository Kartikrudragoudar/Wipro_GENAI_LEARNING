```python
import requests
import json
import os
from dotenv import load_dotenv

# ─── Load environment variables from .env file ────────────────────────────────
load_dotenv(override=True)  # override=True allows .env values to overwrite existing env vars

# ─── Configuration ────────────────────────────────────────────────────────────
API_KEY = os.getenv("ANTHROPIC_API_KEY")
ENDPOINT = "https://llmgw-wp.tekstac.com/v1/chat/completions"

# ─── Headers (same as Postman Headers tab) ────────────────────────────────────
headers = {
    "x-api-key": API_KEY,
    "anthropic-version": "2023-06-01",
    "content-type": "application/json"
}

# ─── Request Body (same as Postman Body → raw → JSON) ─────────────────────────
payload = {
    "model": "global.anthropic.claude-haiku-4-5-20251001-v1:0",
    "max_tokens": 256,
    "temperature": 0.5,
    "system": "You are a helpful assistant.",
    "messages": [
        {
            "role": "user",
            "content": "What is the capital of India?"
        }
    ]
}

# ─── Send POST Request ─────────────────────────────────────────────────────────
response = requests.post(ENDPOINT, headers=headers, json=payload)

assistant_reply = response.json()["choices"][0]["message"]["content"]

# ─── Extract just the assistant's reply ───────────────────────────────────────
assistant_reply = response.json()["choices"][0]["message"]["content"]
print(f"Assistant Reply: {assistant_reply}")
```
