```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np


print("Loading model...")
model = SentenceTransformer("all-MiniLM-L6-v2")
print("Model loaded!\n")


sentences = [
    "GenAI is transforming software development",
    "Artificial Intelligence is changing how developers work",
    "I love playing cricket on weekends"
]

print("Input Sentences:")
for i, sent in enumerate(sentences, 1):
    print(f"{i}. {sent}")
print()


embeddings = model.encode(sentences)

print(f"Embedding dimensions: {embeddings.shape}")
print()


print("First 5 dimensions of each embedding:")
for i, embedding in enumerate(embeddings, 1):
    print(f"Sentence {i}: {embedding[:5]}")
print()


similarity_matrix = cosine_similarity(embeddings)

print("Cosine Similarity Matrix:")
print("=" * 50)
print("       Sent1    Sent2    Sent3")
for i, row in enumerate(similarity_matrix, 1):
    print(f"Sent{i}: {row[0]:.4f}  {row[1]:.4f}  {row[2]:.4f}")
print()


print("Pairwise Similarity Scores:")
print("-" * 50)
for i in range(len(sentences)):
    for j in range(i + 1, len(sentences)):
        sim_score = similarity_matrix[i][j]
        print(f"Sentence {i+1} vs Sentence {j+1}: {sim_score:.4f}")
```
